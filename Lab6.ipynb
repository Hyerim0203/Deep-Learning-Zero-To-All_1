## softmax function

```python
hypothesis = tf.nn.softmax(tf.matmul(x,w)+b)

# tf.matmul(x,w)+b 로 나온 score 을 softmax fuction 을 통해 probability 로 바꿈!
# 하나의 instance 로 나온 확률들의 합은 1

cost = tf.reduce_mean(-tf.reduce_sum(y*tf.log(hypothesis), axis=1)) # 가로로 더한 값. = 하나의 instance 에 대한 cost 이기 때문
                      # axis=1 을 붙여주지 않으면 전체 합이 나옴
                      
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

## model - iris

```python
from sklearn.preprocessing import MinMaxScaler
xy = pd.read_csv("./data-iris.csv",dtype=np.float32)
xy = np.array(xy, dtype=np.float32)

x_train = xy[:,0:-1]
y_train = xy[:,[-1]]
x_train = MinMaxScaler().fit_transform(x_train)

nb_classes = 3

x = tf.placeholder("float", [None, 4])

y = tf.placeholder(tf.int32, [None, 1]) # y_one_hot 의 사용을 위해 int 형식으로 받음!

w = tf.Variable(tf.random_normal([4, nb_classes]), name = "weight")
b = tf.Variable(tf.random_normal([nb_classes]), name = "bias")


logits = tf.matmul(x,w) + b
hypothesis = tf.nn.softmax(logits)

y_one_hot = tf.one_hot(y,nb_classes ) # dtype 이 int 인 것만 받을 수 있음!!!
y_one_hot = tf.reshape(y_one_hot , [-1, nb_classes]) #reshape 에서 None 은 -1 의 형식으로 들어감!

cost_i = tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels = y_one_hot)
# 한개의 인스턴스에 대한 cost의 합이 나옴
cost = tf.reduce_mean(cost_i)

train = tf.train.GradientDescentOptimizer(learning_rate = 0.01).minimize(cost)

prediction = tf.argmax(hypothesis, 1) # axis =1 에 대한 agrmax 값
correct_prediction = tf.equal(prediction, tf.argmax(y_one_hot,1)) # tf.arg_max 와 동일
accuracy = tf.reduce_mean(tf.cast(correct_prediction, 1))

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(1301):
    a,b,_ = sess.run([accuracy, cost, train], feed_dict = {x:x_train, y:y_train})
    if step%400 == 0 :
        print("Step: {:5}\tLoss: {:.3f}\tAcc:{:.2%}".format(step, b, a))
        # 앞에껀 상관 없고, 실수형 소숫점 셋째자리 까지, # 앞에껀 상관없고, %형 소숫점 2째자리 까지
        # :5 칸을 일정하게 유지시켜주는 역할. 

pred = sess.run(prediction, feed_dict={x:x_train})

for a,b in zip(pred, y_train.flatten()): # pred 가 1열만 가지게 되면서 하나의 리스트에 들어간 형태(1차원)가
    #되기 때문에 y_train 또한 flatten 으로 1차원으로 펴주게됨
    print("[{}] Prediction : {} True Y: {}".format(a==int(b), a, int(b)))
    # y_train 의 경우에 float 의 형태이기 때문에 int 로 바꾸어주게 됨!
```

```python
from sklearn.preprocessing import MinMaxScaler
filename_queue = tf.train.string_input_producer(["./data-iris.csv"],shuffle=False, name="filename_queue")

reader=tf.TextLineReader()
key,value = reader.read(filename_queue)

record_defaults = [[0.],[0.],[0.],[0.],[0]]
xy = tf.decode_csv(value, record_defaults = record_defaults)

min_after_dequeue = 10
capacity = min_after_dequeue + 3 * 50
train_x_batch, train_y_batch = tf.train.shuffle_batch([xy[0:-1],xy[-1:]], batch_size = 30,
                                                     capacity = capacity, min_after_dequeue = min_after_dequeue)

nb_classes = 3

x = tf.placeholder("float", [None, 4])
y = tf.placeholder(tf.int32, [None, 1])

w = tf.Variable(tf.random_normal([4, nb_classes]), name = "weight")
b = tf.Variable(tf.random_normal([nb_classes]), name = "bias")


logits = tf.matmul(x,w) + b
hypothesis = tf.nn.softmax(logits)

y_one_hot = tf.one_hot(y,nb_classes )
y_one_hot = tf.reshape(y_one_hot , [-1, nb_classes]) #reshape 에서 None 은 -1 의 형식으로 들어감!

cost_i = tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels = y_one_hot)
# 한개의 인스턴스에 대한 cost의 합이 나옴
# -tf.reduce_sum(tf.log(hypothesis)*y_one_hot) 와 동일함
cost = tf.reduce_mean(cost_i)

train = tf.train.GradientDescentOptimizer(learning_rate = 1e-4).minimize(cost)

prediction = tf.argmax(hypothesis, 1)
correct_prediction = tf.equal(prediction, tf.argmax(y_one_hot,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, 1))

sess = tf.Session()
sess.run(tf.global_variables_initializer())

coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)

for step in range(2001):
    x_train, y_train = sess.run([train_x_batch, train_y_batch])
    x_train = MinMaxScaler().fit_transform(x_train)
    c_val,_ = sess.run([cost, train], feed_dict = {x:x_train, y:y_train})
    if step%200 == 0:
        print(c_val)
```
