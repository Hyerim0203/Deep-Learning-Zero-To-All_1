- text : "hihello" => "o" 다음 문자가 없기 때문에 hihell 까지만 input
- unique chars(vocabulary, voc): h, i, e, l, o
- voc index= h:0, i:1, e:2, 1:3, o:4

- sequence : 6
- hidden_layer : 5
- batch : 1
- input_dimension : 5

```python
import tensorflow as tf
import numpy as np
import pprint
sess=tf.InteractiveSession()
```

## Data creation

```python
dic_char = {}
for idx, i in enumerate(list("hielo")):
    dic_char[i]=idx
x_data = list("hihell")
x_data = [dic_char[i] for i in x_data]
x_data = tf.reshape(tf.one_hot(x_data, len(dic_char)), [1,-1,5])
y_data = np.array([[dic_char[i] for i in list("ihello")]])
print(y_data)
print(x_data.eval())
```

```python
x_data=np.array(x_data.eval())
```

```python
dic_char_num = {0:"h",1:"i",2:"e",3:"l",4:"o"}
```

## Feed to RNN

```python
#parameters
with tf.variable_scope("hife13ee33"):
    sequence_length = 6
    hidden_size = 5
    input_dim = 5
    batch_size=1

    x = tf.placeholder(tf.float32, [None, sequence_length, input_dim])
    y = tf.placeholder(tf.int32, [None, sequence_length])

    cell = tf.contrib.rnn.BasicLSTMCell(hidden_size, state_is_tuple=True)
    #  초기 output의 state를 정함
    initial_state = cell.zero_state(batch_size, tf.float32)
    # initial_state의 shape는 [batch_size, hidden_layer]
    outputs, states = tf.nn.dynamic_rnn(cell, x, initial_state=initial_state,
                                       dtype=tf.float32)
    weights = tf.ones([batch_size, sequence_length])
    
    sequence_loss = tf.contrib.seq2seq.sequence_loss(
                logits=outputs, targets=y, weights=weights) #각 batch 의 loss
    loss = tf.reduce_mean(sequence_loss) # 모든 batch의 loss
    train = tf.train.AdamOptimizer(learning_rate=0.1).minimize(loss)
    
    prediction = tf.argmax(outputs, axis=2)
    
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        for i in range(2000):
            l, _ = sess.run([loss,train], feed_dict={x:x_data,
                                                    y:y_data})
            result = sess.run(prediction, feed_dict={x:x_data})
            print(i, "loss", l, "prediction", result, "true", y_data)
            result_str = [dic_char_num[c] for c in np.squeeze(result)] # 각숫자해당문자열
            print("prediction", "".join(result_str)) #문자열 합침
    
```
