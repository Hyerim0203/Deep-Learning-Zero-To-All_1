## Hypothesis not using matrix 

### *H(x1,x2,x3) = x1w1+ x2w2 + x3w3 *

```python
import tensorflow as tf

x1_data=[73,93,89,86,73]
x2_data=[80,88,91,98,66]
x3_data=[75,93,90,100,70]
y_data=[152,185,180,196,142]

# placeholders for a tensor that will be always fed
x1=tf.placeholder(tf.float32,shape=[None]) #1차원이고 원소 개수는 명시하지 않아주었음
x2=tf.placeholder(tf.float32)
x3=tf.placeholder(tf.float32)
# placeholder 는 반드시 dtype 을 명시해주어야 함!

y=tf.placeholder(tf.float32, shape=[None])

w1=tf.Variable(tf.random_normal([1]),name="weight1")
w2=tf.Variable(tf.random_normal([1]),name="weight2")
w3=tf.Variable(tf.random_normal([1]),name="weight3")

b=tf.Variable(tf.random_normal([1]),name="bias")

hypothesis=x1*w1+x2*w2+x3*w3+b

# cost/Loss function
cost=tf.reduce_mean(tf.square(hypothesis-y))

#Minimize Need a very small learning rate for this data set
optimizer=tf.train.GradientDescentOptimizer(learning_rate=1e-5) # e는 1/10..?
train=optimizer.minimize(cost)

# Launch the graph in a session
sess=tf.Session()
sess.run(tf.global_variables_initializer())
for step in range(2001):
    cost_val, hy_val,_ = sess.run([cost,hypothesis,train],
                                  feed_dict={x1:x1_data,x2:x2_data,x3:x3_data,y:y_data})
    if step%10==0:
        print(cost_val,hy_val)
```

## Hypothesis using matrix

```python
x_data=[[73.,80.,75.],[93.,88.,93.],[89.,91.,90.],[86.,98.,100.],[73.,66.,70.]]
y_data=[[152.],[185.],[180.],[196.],[142.]]

# placeholders for a tensor that wiil be always fed.
x = tf.placeholder(tf.float32, shape=[None,3]) # 데이터의 개수는 정하지 않고 feature 은 3개
# 행은 정하지 않고 열 개수만 고정 # 3열만 고정
y = tf.placeholder(tf.float32, shape=[None,1])

w = tf.Variable(tf.random_normal([3,1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")

# Hypothesis
hypothesis = tf.matmul(x,w) + b # x와 w를 일반 곱이 아닌 행렬곱을 해주어야 함!!!

# Simplified cost/Loss function
cost=tf.reduce_mean(tf.square(hypothesis-y))
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train=optimizer.minimize(cost)

# Launch the graph in a session
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val,_ = sess.run([cost,hypothesis,train], feed_dict={x:x_data,y:y_data})
    if step%500==0:
        print(step,"cost:",cost_val,"prediction",hy_val)
    # hypothesis 또한 노드에 얽혀있는 연산이기 때문에 반드시 run 을 해준 뒤 값을 반환해야 함!
```

```python
import tensorflow as tf
filename_queue = tf.train.string_input_producer(["./data-01-test-score.csv"], shuffle=False, name="filename_queue")

reader = tf.TextLineReader()
key, value = reader.read(filename_queue)

# Default values, in case of empty columns. Also specifies the type of the
# decorded result.
record_defaults = [[0.] for i in range(4)] # 열들의 형식을 정함
xy = tf.decode_csv(value, record_defaults=record_defaults) # 해당 파일이름들의 파일 내용을 csv 로 불러옴

# collect batches of csv in
train_x_batch, train_y_batch = tf.train.batch([xy[0:-1],xy[-1:]], batch_size=2) # xy[-1] 하게 되면 또 마음대로 하나로바꿔줌
# xy[-1:] 해줘야지 원래 형식 그대로 가져옴 
# batch 에서의 열 슬라이싱은 [-1] 이런 식으로 안에 integer 가 들어오거나 0:3 3: 형식의 슬라이싱만 가능
# xy[[-1]] 불가능
    
# placeholders for a tensor that will be always fed
x = tf.placeholder(tf.float32, shape=[None,3])
y = tf.placeholder(tf.float32, shape=[None,1])

w= tf.Variable(tf.random_normal([3,1]),name="weight")
b= tf.Variable(tf.random_normal([1]), name="bias")

# Hypothesis
hypothesis = tf.matmul(x,w) + b

# Simplified cost/Loss Function
cost = tf.reduce_mean(tf.square(hypothesis-y))

# MInimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train=optimizer.minimize(cost)

#Launch the graph in a session
sess=tf.Session()
sess.run(tf.global_variables_initializer())

# Start populating the filename queue.
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)
# 큐를 해당 모델에 실행시킴

for step in range(2001):
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch])  # 배치를 실행시킴
    cost_val, hy_val,_ = sess.run([cost,hypothesis, train], feed_dict={x:x_batch, y:y_batch})
    # 조금씩 뽑아온 데이터로 해당 모델을 실행하면서 모델을 업데이트 시킴
    if step%200==0:
        print(step, cost_val, hy_val)
        print(x_batch, y_batch)
        
coord.request_stop()
coord.join(threads)

# queue 를 멈춤

```
